```python
# calculando as estatísticas para cada combinação de qualidade (5, 6, 7) e preditor (11 características)

# Valores de qualidade que deseja analisar
qualidades_desejadas = [5, 6, 7]  # Qualidades escolhidas (pode alterar conforme necessário)

# Pode ser um loop para cada preditor (característica)
valores = ["fixed acidity", "volatile acidity", "citric acid", "residual sugar", "chlorides", "free sulfur dioxide", "total sulfur dioxide", "density", "pH", "sulphates", "alcohol"]

# Inicialize listas para armazenar os resultados
resultados = []

for qualidade in qualidades_desejadas:
    for preditor in valores:
        # Filtrar o conjunto de dados para a qualidade selecionada
        tintos_qualidade_escolhida = tintos[tintos['quality'] == qualidade]

        # Calcula a média condicional à classe para Tintos
        media_tintos = tintos_qualidade_escolhida[preditor].mean()

        # Calcula o desvio padrão condicional à classe para Tintos
        desvio_padrao_tintos = tintos_qualidade_escolhida[preditor].std()

        # Calcula a assimetria condicional à classe para Tintos
        assimetria_tintos = tintos_qualidade_escolhida[preditor].skew()

        # Adiciona os resultados à lista
        resultados.append([qualidade, preditor, media_tintos, desvio_padrao_tintos, assimetria_tintos])

# Cria um DataFrame com os resultados
dados = pd.DataFrame(resultados, columns=["Qualidade", "Preditor", "Média Tintos", "Desvio Padrão Tintos", "Assimetria Tintos"])

# Exibe o DataFrame
print(dados)
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import pandas as pd
import matplotlib.pyplot as plt


# Una os dois conjuntos de dados
dados_combinados = pd.concat([tintos, brancos], ignore_index=True)

# Suponha que seu conjunto de dados esteja em um DataFrame chamado 'df'
X = dados_combinados.drop('Tipo', axis=1)
# Normalizar os dados, se necessário (opcional)

scaler = StandardScaler(with_std=True, with_mean=True)
dados_padronizados = scaler.fit_transform(X)

# Aplicar a PCA
vinhos_pca = PCA()
vinhos_pca.fit_transform(dados_padronizados)

pontuacao = vinhos_pca.transform(dados_padronizados)

#como estão dividas as proporções
vinhos_pca.explained_variance_ratio_

# Adicione a coluna 'Tipo' de volta aos dados transformados
dados_transformados = pd.DataFrame(pontuacao, columns=[f'PC{i + 1}' for i in range(pontuacao.shape[1])])
dados_transformados['Tipo'] = dados_combinados['Tipo']

# Defina cores para tintos e brancos
cores = {'Tinto': 'red', 'Branco': 'magenta'}

sns.set(style="darkgrid")
i, j = 0, 1 # which components
fig, ax = plt.subplots(1, 1, figsize=(8, 8))
for tipo, cor in cores.items():
    mask = dados_transformados['Tipo'] == tipo
    ax.scatter(dados_transformados.loc[mask, f'PC{i + 1}'], dados_transformados.loc[mask, f'PC{j + 1}'], label=tipo, color=cor)

ax.set_xlabel('PC%d' % (i+1))
ax.set_ylabel('PC%d' % (j+1))
for k in range(vinhos_pca.components_.shape[1]):
    ax.arrow(0, 0, vinhos_pca.components_[i,k], vinhos_pca.components_[j,k])
    ax.text(vinhos_pca.components_[i,k],
            vinhos_pca.components_[j,k],
            dados_combinados.columns[k])


scale_arrow = s_ = 10
fig, ax = plt.subplots(1, 1, figsize=(8, 8))
#ax.scatter(pontuacao[:,0], pontuacao[:,1])
for tipo, cor in cores.items():
    mask = dados_transformados['Tipo'] == tipo
    ax.scatter(dados_transformados.loc[mask, f'PC{i + 1}'], dados_transformados.loc[mask, f'PC{j + 1}'], label=tipo, color=cor)


ax.set_xlabel('PC%d' % (i+1))
ax.set_ylabel('PC%d' % (j+1))
for k in range(vinhos_pca.components_.shape[1]):
    ax.arrow(0, 0, s_*vinhos_pca.components_[i,k], s_*vinhos_pca.components_[j,k])
    ax.text(s_*vinhos_pca.components_[i,k],
            s_*vinhos_pca.components_[j,k],
            dados_combinados.columns[k])




# Variância explicada por cada componente
explained_variance_ratio = vinhos_pca.explained_variance_ratio_


# Crie um gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio, alpha=0.5, align='center')
plt.step(range(1, len(explained_variance_ratio) + 1), np.cumsum(explained_variance_ratio), where='mid')
plt.xlabel('Componente Principal')
plt.ylabel('Variância Explicada')
plt.title('Variância Explicada por Componente Principal')
plt.grid(True)
plt.show()

```
